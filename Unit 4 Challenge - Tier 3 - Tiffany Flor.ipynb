**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Create function to calculate ratio of 2020:2000
# example of what d will be: dfg[dfg['London_Borough']=='Camden']
def create_price_ratio(d):
    y2000 = float(d['Average_price'][d['Year']==2000])
    y2020 = float(d['Average_price'][d['Year']==2020])
    ratio = [y2020/y2000]
    return ratio
```

```python
# Test it out on Camden
create_price_ratio(dfg[dfg['London_Borough']=='Camden'])
```

```python
# Apply this function to all Buroughs
# Create dictionary for all of the ratios created
final = {}
```

```python
# Create for loop to iterate over all of the unique boroughs
for b in dfg['London_Borough'].unique():
    # creates argument to pass through create_price_ratio: dfg[dfg['London_Borough']=='Camden']
    borough = dfg[dfg['London_Borough'] == b]
    # make new entry in dictionary for b
    final[b] = create_price_ratio(borough)

# print our new dictionary of ratios for each borough
print(final)
```

```python
# Create a dataframe from the borough price ratios
df_ratios = pd.DataFrame(final)
df_ratios.head()
```

```python
# Transpose this dataframe to get boroughs as rows (use .T shortcut)
df_ratios_T = df_ratios.T
# Check out what we made
df_ratios_T.head()
```

```python
# Reset index so it is easier to work with
df_ratios = df_ratios_T.reset_index()

# Print what we have
df_ratios.head()
```

```python
# Rename columns - 'index':'Borough', 0:'Price'
# inplace = True reassigns to the same dataframe to save it
df_ratios.rename(columns = {'index':'Borough', 0:'2020'}, inplace=True)

# Print what we have
df_ratios.head()
```

```python
# Order the top 10 from greatest increase descending in values
top15 = df_ratios.sort_values(by='2020', ascending=False).head(15)
print(top15)
```

```python
# Plot this as a bar graph
ax = top15[['Borough','2020']].plot(kind='bar', legend=None)

ax.set_xticklabels(top15.Borough)

ax.set(title='Change in Housing Price',
      ylabel='Percent Change from 2000 to 2020',
      xlabel='London Borough')
```

```python
# Let's check out the bottom 15 to see if every borough had a percent increase over the past 10 years
bottom15 = df_ratios.sort_values(by='2020', ascending=True).head(15)

ax = bottom15[['Borough','2020']].plot(kind='bar', legend=None)

ax.set_xticklabels(bottom15.Borough)

ax.set(title='Change in Housing Price',
      ylabel='Percent Change from 2000 to 2020',
      xlabel='London Borough')
```
